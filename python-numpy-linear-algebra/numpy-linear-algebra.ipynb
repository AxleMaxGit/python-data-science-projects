# Solving Linear Equations

## Q: Find the equation for the line joining points (1,1) and (6,9)

```python
import numpy as np
```

A general line is in the frm m(x) + b = y

in linear algebra is represented asd [x, 1] = y  

[note:the 1 represents "a coefficient of 1 for the constant b"]

So for point (6,9)
the equation becomes [6, 1] = 9

#Solving for 2 points (1,1) & (6,9)

#becomes ...

#Equation is y = mx + b

#RHS is the coefficients ot the RHS

for point 1 coefficents are, m=1, b=1

for point 2 coefficents are, m=6, b=1

```python
RHS = np.matrix([
    [1, 1],
    [6, 1]
])

print(RHS)
```

for point 1 LHS value is y=1

for point 2 LHS value is y=9

```python
LHS = np.matrix([
    [1], 
    [9]
])

print(LHS)
```

## Linear Algrbraic representation of the system of equations for points (1,1) & (6,9) is ...


```python
#Before
from IPython.display import Image
Image(filename='before.png') 
```

## Rearrange using algebra to solve for m & b

```python
#after
Image(filename='after.png')
```

```python
#Inverse the RHS matrix
RHS_INV = np.linalg.inv(RHS) 
RHS_INV
```

## Gives m=1.6 & b=-0.6

```python
np.dot(RHS_INV, LHS)
```

## Equation for the line joining point (1, 1) & (6,9) is therefore ... y = 1.6x - 0.6
